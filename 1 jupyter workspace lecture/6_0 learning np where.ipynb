# **np.where**

```python
import numpy as np

arr = np.array([4,5,6,7,8,9,11,0,25])

# x = ['HIGH','HIGH','HIGH','HIGH','HIGH','HIGH','HIGH','HIGH','HIGH']
# y = ['LOW','LOW','LOW','LOW','LOW','LOW','LOW','LOW','LOW']

new_arr = np.where(
    arr > 5,
    'HIGH',
    'LOW'
)

print(new_arr)
```

```python
arr = np.array([4,5,6,7,8,9,11,0,25])

x = ['HIGH','HIGH','HIGH','HIGH','HIGH','HIGH','HIGH','HIGH','HIGH']
y = ['LOW','LOW','LOW','LOW','LOW','LOW','LOW','LOW','LOW']

new_arr = np.where(
    arr > 5,
    x,
    y
)

print(new_arr)
```

```python
arr = np.array(np.random.randint(0,21,12))

new_arr = np.where(
    arr % 2 == 0,
    'E',
    'O'
)

print(arr)
print(new_arr)
print()

for num, char in zip(arr, new_arr):
    print(f"{num} - '{char}'")

print()
print(", ".join(f"{n}-'{c}'" for n, c in zip(arr, new_arr)))
```
