## .values

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [40, 50, 60]
})

a1 = df.iloc[0]
print(a1)
print(type(a1), '\n')

a2 = df.iloc[0].values
print(a2)
print(type(a1))
```

```python
a3 = df.values
print(a3, '\n')

a4 = df.values[0]
print(a4)
```
