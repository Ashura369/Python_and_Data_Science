```python
import numpy as np
np.random.seed(10)

# 1. The Center (Mean)
# Let's say we want data centered around x=10, y=20
mean = [10, 20]   

# 2. The Shape (Covariance)
# [[1, 0], [0, 1]] means a simple circle with variance 1
cov = [[1, 0],          # this is an identity metrics
       [0, 1]]
# 3. Generate 5 points
data = np.random.multivariate_normal(mean, cov, 5)
# remember 'multivariate_normal' takes two inputs one for mean and another for covariance




print("\nGenerated Data (5 points):")
print(data)
```
