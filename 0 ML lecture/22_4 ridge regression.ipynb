# **Ridge Regression using Gradient Descent**

> So far we have been learning ridge reg on linear reg. Now we will be learning ridge reg using Gradient descent.

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, SGDRegressor      
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_diabetes


X,y = load_diabetes(return_X_y=True)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)

# making prediction using SGDRegressor
reg = SGDRegressor(penalty='l2', max_iter=500, eta0=0.1, learning_rate='constant', alpha=0.001)
# since you are using penaly = 'l2', it automatically becomes SGDRegressor along with ridge regression. Bcoz, l2 represents 'ridge'
# refer 20_4_2 for all the parameters of SGDRegressor
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)

print("R2 score : ",r2_score(y_test,y_pred), '\n')
print(reg.coef_)
print(reg.intercept_)
```

```python
# making the same prediction using Ridge regression 

reg = Ridge(alpha=0.001, max_iter=500,solver='sparse_cg')
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)

print("R2 score",r2_score(y_test,y_pred))
print(reg.coef_)
print(reg.intercept_)

```
