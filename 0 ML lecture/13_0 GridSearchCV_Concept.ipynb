# Part 1: What is GridSearchCV? 

Imagine you are trying to crack a safe that has a 2-digit combination lock.

*   The first digit can be **1, 2, or 3**.
*   The second digit can be **A, B, or C**.

To find the correct combination, you would try **every possible pair**:
1.  (1, A)
2.  (1, B)
3.  (1, C)
4.  (2, A)
... and so on.

**GridSearchCV** does exactly this for Machine Learning models.

### 1. The "Grid" (The Combinations)
Machine Learning models have settings called **Hyperparameters** (e.g., "how deep should the tree grow?"). You can't learn these from the data; you have to set them beforehand.
Instead of guessing, you define a "grid" of options:
*   **Max Depth:** `[None, 10, 20]`
*   **Criterion:** `["gini", "entropy"]`

The "Grid" is just the table of all possible combinations of these settings.

### 2. The "CV" (Cross-Validation)
If you just check if a combination works well on your training data, you might be cheating (overfitting). You need to know if it works on *new* data.
**Cross-Validation** splits your data into parts (folds), usually 5.
*   It trains on 4 parts.
*   It tests on the 1 remaining part.
*   It repeats this 5 times so every part gets a chance to be the test set.
*   The final score is the **average** of these 5 runs.

### 3. The "Search"
The computer automatically:
1.  Picks one combination from your grid.
2.  Runs Cross-Validation (trains 5 times).
3.  Calculates the average score.
4.  Moves to the next combination.
5.  At the end, it tells you which combination had the highest score.
