```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
# Sample data
data = {
    'Age': [53, 25, 63, 39, 22, 59, 41, 55, 30, 47,
            33, 61, 45, 27, 49, 65, 36, 43, 51, 57],
    
    'Salary': [51000, 28000, 67000, 42000, 23000, 59000, 45000, 62000, 31000, 52000,
               36000, 64000, 47000, 29000, 55000, 68000, 40000, 46000, 48000, 60000],
    
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female',
               'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female']
}

df = pd.DataFrame(data)
print("Original Data:")
print(df.head())

# Normalization (only on numeric columns)
scaler = MinMaxScaler()
df_normalized = df.copy()           
df_normalized[['Age', 'Salary']] = scaler.fit_transform(df[['Age', 'Salary']])
# normalization dont work on text data, hence we first copied the data and then used normalization only on Age and Salary column, which were completely numerical

print("\nAfter Normalization:")
print(df_normalized.head())
```
