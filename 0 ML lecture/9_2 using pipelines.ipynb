## Using Pipelines

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline, make_pipeline            # used to make pipeline
from sklearn.feature_selection import SelectKBest, chi2
```

```python
df = pd.read_csv('train.csv')
df.head(3)
```

```python
# dropping unrequired columns
df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])

# checking which column has the missing values
print(f"MISSING VALUES : \n{df.isna().sum()}\n") # Age and Embarked has the missing values

x_train, x_test, y_train, y_test = train_test_split(df.drop(columns='Survived'), df['Survived'], test_size=0.2, random_state=0)

# using transformer
# imputation transformer
trf1 = ColumnTransformer([('impute_age', SimpleImputer(), [2]), ('impute_embarked', SimpleImputer(strategy='most_frequent'), [6])], 
                         remainder='passthrough')

# one hot encoding transformer
trf2 = ColumnTransformer([('ohe_sex_embarked', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [1,6])], 
                         remainder='passthrough')       # if you dont put 'passthrough' rest of the columns would have been dropped (deleted)

# scaling transformer
trf3 = ColumnTransformer([('scale', MinMaxScaler(), slice(0,10))])      # from 0 to 9, 10 being explict

# feature selection
trf4 = SelectKBest(score_function=chi2, k=8)        # using only top 8 columns for feature selection to make prediction

# training the model
trf5 = DecisionTreeClassifier()


```
