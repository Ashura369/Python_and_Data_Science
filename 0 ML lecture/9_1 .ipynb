### Using Pickle

The code down below is the production code, which you will be using while working in a corporate. And the code we wrote in 9_0 was the code that the data scientist has to write on his computer, and then upload the encoders and predictions on the file. 

```python
import pickle
import numpy as np
import pandas as pd
```

```python
ohe_sex = pickle.load(open('models/ohe_sex.pkl', 'rb'))
ohe_embarked = pickle.load(open('models/ohe_embarked.pkl', 'rb'))
clf = pickle.load(open('models/clf.pkl', 'rb'))
col = pickle.load(open('models/col.pkl', 'rb'))

# making new user input
# Pclass/gender/age/SibSp/Parch/Fare/Embarked
test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'], dtype=object).reshape(1, 7)


# using ohe hot encoding on sex and embarked which is imported from 9_0 file
# this 'sex' holds both the columns for male and female
test_input_sex = ohe_sex.transform(test_input[:, 1].reshape(1, 1))
test_input_embarked = ohe_embarked.transform(test_input[:, -1].reshape(1, 1))


test_input_age = test_input[:, 2].reshape(1, 1)


# concanating all the columns into one column
test_input_transformed = np.concat((test_input[:, [0,3,4,5]], test_input_age, test_input_sex, test_input_embarked), axis=1)
test_input_transformed

```

```python
# converting into data frame and naming them with columns

# we are using 'col' which is imported from 9_0
test_input_transformed = pd.DataFrame(test_input_transformed, columns=col)
test_input_transformed
```

```python
# making prediction
clf.predict(test_input_transformed)
# as the ans is showing 0, that means the given person will die
```
