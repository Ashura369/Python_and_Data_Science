### Summary of Bias-Variance Tradeoff

To determine if a model has a "good" bias-variance tradeoff, we look for a balance where both training and testing errors are minimized and close to each other.

| Scenario | Training Error | Test Error | Status |
| :--- | :--- | :--- | :--- |
| **High Bias** | 25% (High) | 27% (High) | **Underfitting**: The model is too simple to learn the data. |
| **High Variance** | 1% (Low) | 20% (High) | **Overfitting**: The model memorized training data but fails on new data. |
| **Good Tradeoff** | 8% (Low) | 10% (Low) | **Balanced**: The model generalizes well to unseen data. |

**The Goal:** We aim for the "Sweet Spot" where the total error (BiasÂ² + Variance + Irreducible Error) is at its lowest point.



So to do a good trade off, you must focus on a low bias and also a low variance.
