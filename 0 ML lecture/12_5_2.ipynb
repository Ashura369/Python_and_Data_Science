# Imputation by chained equation of missing value

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression



import warnings
warnings.filterwarnings('ignore')
```

```python
df = np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)
```

```python
df = df.iloc[:, [0,1,2]]

# assigning NaN values to different cells
df.iloc[1,0] = np.nan
df.iloc[3,1] = np.nan
df.iloc[-1,-1] = np.nan

# assigning mean values to all the NaN cells
df_imputed = df.copy()
for col in df_imputed.columns:
    df_imputed[col] = df_imputed[col].fillna(df_imputed[col].mean())



df_imputed
```

```python
# making a dictionary of missing numbers 
cols_with_missing = {
    0 : (1,0),
    1 : (3,1),
    2 : (-1,-1)
}


# using for loop
iterations = 5 
for i in range(iterations):
    print(f"ITERATION : {i+1}")

    for col_idx, (row_idx, _) in cols_with_missing.items():
        temp_df = df_imputed.copy()
        
        # assigning NaN values
        temp_df.iloc[row_idx, col_idx] = np.nan
        
        x_train = temp_df.dropna().drop(temp_df.columns[col_idx], axis=1)
        y_train = temp_df.dropna().iloc[:,col_idx]

        x_test = temp_df.iloc[[row_idx]].drop(temp_df.columns[col_idx], axis=1)

        lr = LinearRegression()
        lr.fit(x_train, y_train)
        y_test_pred = lr.predict(x_test)[0]

        df_imputed.iloc[row_idx, col_idx] = y_test_pred
        
        print(f"Updated {df.columns[col_idx]} at row {df.index[row_idx]}: {y_test_pred:.2f}")
    print("\nResult after Iteration:\n", df_imputed)

```
